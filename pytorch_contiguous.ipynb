```python
import numpy as np
from numpy import random
import torch
```

```python
test_1 =np.random.randint(1, 10, (5, 6, 3, 2))
test_1 = torch.from_numpy(test_1)
print(test_1)
```

```python
test_2 = test_1[:, 2:-2, :, :2].contiguous()
print('test_2.shape : ', test_2.shape)
print('test_2 : ', test_2)
```

contiguous() makes x and y don't share the same memory.

```python
test_2[0][0][0] = 1000000
```

```python
print(test_1)
```

If we don't use contiguous, the test_3 value will change as we change test_4 value.

```python
test_3 = np.random.randint(1, 10, (6, 2))
test_4 = test_3[2:-2, :2]
```

```python
print('test_3 : ', test_3)
```

```python
print('test_4 : ', test_4)
```

```python
test_4[0][0] = 1000000
print('test_3 : ', test_3)
```

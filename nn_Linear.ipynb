```python
import torch
import numpy as np
import torch.nn as nn
import torch.nn.functional as F
```

```python
x = torch.randn(2, 4, 7) # batch number : 2, frame length : 4, channel : 7
print("x.shape : ", x.shape, '\n')
print("x : ", x)
```

```python
layer = nn.Linear(in_features = 7, out_features = 10)
```

```python
print("layer.weight.shape : ", layer.weight.shape, '\n')
print("layer.weight : ", layer.weight)
```

```python
print("layer.bias.shape : ", layer.bias.shape, '\n')
print("layer.bias : ", layer.bias)
```

```python
out = layer(x)
print("out.shape : ", out.shape, '\n')
print("out : ", out)
```

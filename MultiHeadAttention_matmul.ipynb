```python
import torch
import numpy as np
import torch.nn as nn
import torch.nn.functional as F
```

```python
# for testing multihead attention

x = torch.randint(10, (2, 3, 4, 6)) 
# batch size : 2
# head number : 3
# row number : 4 (frame number)
# column number : 6 (channel for one head)


y = torch.randint(10, (2, 3, 4, 6)) 
# batch size : 2
# head number : 3
# row number : 4 (frame number)
# column number : 6 (channel for one head)

y = y.transpose(-2, -1)
```

```python
print("x.shape :", x.shape, '\n')
print("x : ")
print(x)
```

```python
print("y.shape :", y.shape, '\n')
print("y : ")
print(y)
```

```python
result = torch.matmul(x, y)
print("result.shape :", result.shape, '\n')
print("result : ")
print(result)
```
